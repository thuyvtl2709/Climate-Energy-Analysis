#TOP 3 countries produces the most C02 emission per capita in 2022

```python
import pandas as pd
import plotly.graph_objs as go
```

```python
result2022Plot_country['CO₂ emissions per capita'] = result2022Plot_country['Annual CO₂ emissions'] / result2022Plot_country['2022 Population']
```

```python
result2022Plot_country
```

```python
sorted_df = result2022Plot_country.sort_values(by='CO₂ emissions per capita', ascending=False)
```

```python
sorted_df
```

```python
sorted_df["CO₂ emissions per capita"] = sorted_df["CO₂ emissions per capita"].round(2)
```

```python
sorted_df
```

```python
top_3_countries = sorted_df.head(10)
```

```python
fig = go.Figure()
```

```python
fig.add_trace(go.Bar(
    x=top_3_countries.index,  # Tên các quốc gia
    y=top_3_countries['CO₂ emissions per capita'],
    marker_color='#636EFA',  # Màu của cột
    text=top_3_countries['CO₂ emissions per capita'],  # Hiển thị số liệu trên cột
    textposition='auto',  # Vị trí của số liệu
))
```

```python
fig.update_layout(
    title="Top 10 countries with highest CO2 emissions per capita in 2022",
    xaxis_title="Country",
    yaxis_title="CO2 emissions per capita")
```
